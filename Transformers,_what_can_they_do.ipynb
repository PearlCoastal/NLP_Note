```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")

classifier(
    ["let's go.", "I don't understand what you are saying."]
    )
```

```python
generator = pipeline("text-generation")

generator("At this summer, I will")
```

```python
generator = pipeline("text-generation", model="distilgpt2")

generator("At this summer, I will",

          max_length = 30,

          num_return_sequence=2,
)
```

```python
unmasker = pipeline("fill-mask")

unmasker("I will teach you how to say the <mask> word.", top_k = 2)
```

```python
ner = pipeline("ner", grouped_entities=True)

ner("My name is FEIFEI, I live in Nomi-city and I am studying NLP at JAIST.")
```

```python
question_answerer = pipeline("question-answering")

question_answerer(question= "how old are you",
                  context= "My name is FEIFEI and I am 26 years old.")
```

```python
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")

translator("Ce cours est produit par Hugging Face.")
```
